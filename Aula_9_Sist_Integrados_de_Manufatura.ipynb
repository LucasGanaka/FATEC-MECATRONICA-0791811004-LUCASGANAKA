```
#Introdução ao Pandas

import pandas as pd

#para utilizar as feramentas localmente
```

```
#Estruturas de dasos com Pandas:
#Series - uma coluna de dados 
#DataFrames - tabelas de dados  
```

```
#Series
notas = [10, 8, 9, 4, 5, 6.7, 8, 7, 6, 5]
alunos = ['charmander', 'Bulbassaur', 'Squirtle', 'Wartatle', 'Ivysaur', 'Charmander', 'Blastoise', 'Charizard', 'Venossaur', 'Pikachu']

#Criar serie de dados para utilização
serie_dados = pd.Series(notas, index=alunos)
```

```
print(serie_dados)
print(serie_dados[0])

#Fazendo uma verificação se existe no indice
if 'Venossaur' in serie_dados:
  print(serie_dados['Venossaur'])
```

```
#DataFrame
dados_herois = { "nome" : ['Batman', 'Homem de Ferro', 'Homem Aranha', 'Antman', 'Thor', 'Constantine'],
                'quantidade_de_filmes':[12, 10, 9, 5, 7, 1],
                'historia_tragica':[1,1,1,0,0,1],
                'arquiinimigo':[100, 3, 5, 1, 1, 1],
                'casa':['DC', 'Marvel', 'Sony', 'Marvel', 'Marvel', 'Vertigo']}
#Até aqui um dicionário comum
print(dados_herois)
```

```
#Cria um dataframe (df) dos herois
df_herois = pd.DataFrame(dados_herois, index=dados_herois['nome'])
print(df_herois)
```

```
#conhecer os tipos que existem no df
print(df_herois.dtypes)
```

```
print(df_herois['casa'])
```

```
#Estatistica com dados de tendencia central
print("Quantidade de filmes:")
print("Media:", df_herois['quantidade_de_filmes']. mean())
print("Moda:", df_herois['quantidade_de_filmes'].mode().values)
print("Mediana:", df_herois['quantidade_de_filmes'].median())
```

```
#Analise Gráfica - Bloxplot

boxplot = df_herois.boxplot(column=['quantidade_de_filmes'], showmeans=True)
```

```
#Analisar conjunto de colunas 

df_dados_numericos = df_herois[["quantidade_de_filmes", "historia_tragica", "arquiinimigo"]]
print(df_dados_numericos)
```

```
df_dados_numericos.describe()
```

```
df_dados_numericos.boxplot()
```
